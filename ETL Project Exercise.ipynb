## Extraction from 2 CSV files

````python
#import required dependencies here:
import pandas as pd
import datetime
from sqlalchemy import create_engine
#from config import db_password
````

````python
#Read all_years_o3.csv into df1 and display the first 5 rows
df1 = pd.read_csv("all_years_o3.csv")
df1.head()
````

````python
df1.info()
````

````python
#Read all_years_pm25.csv into df2 and display the first 5 rows
df2 = pd.read_csv('all_years_pm25.csv')
df2.head()
````

# Transform

````python
#drop last 3 columns on df1 and display the first 5 rows.
df1 = df1.drop(labels=['min (ppb)', 'max (ppb)', 'median (ppb)'], axis=1)

df1.head()
````

````python
#rename the count columns in df1 to 'Count_o3' and show first 5rows
df1 = df1.rename(columns = {'count':'Count_o3'})

df1.head()
````

````python
#delete specie in df1 and whow first 5rows
df1 = df1.drop(labels = 'Specie', axis=1)

df1.head()
````

````python
df2.info()
````

````python
#drop last 3 columns on df2 and display the first 5 rows.
df2 = df2.drop(labels= ['min (ug/m3)', 'max (ug/m3)', 'median (ug/m3)'], axis = 1 )
df2.head()
````

````python
#rename the count columns in df2 to 'Count_o3' and show first 5rows
df2 = df2.rename(columns ={'count':'Count_o3'})
df2.head()
````

````python
#delete specie in df2 and whow first 5rows
df2 = df2.drop(labels = 'Specie', axis=1)

df2.head()
````

````python
#merge the two dfs into one on date, country, city and show the first 5rows
merged_df = pd.merge(df1, df2, on=['Date', 'Country', 'City'])

merged_df.head()
````
